This notebook clones the repo into `/content/parameter-golf`, installs dependencies, downloads challenge data, runs record preflight, and provides both a smaller Colab pilot run and the autonomous loop start command.


```python
%%bash
set -euo pipefail
WORKDIR=/content/parameter-golf
REPO_URL=https://github.com/aryan-cs/parameter-golf.git
if [ ! -d "$WORKDIR/.git" ]; then
  git clone "$REPO_URL" "$WORKDIR"
else
  git -C "$WORKDIR" pull --ff-only
fi
cd "$WORKDIR"
git rev-parse HEAD
```

```python
%%bash
set -euo pipefail
cd /content/parameter-golf
python3 -m pip install -U pip
python3 -m pip install -r research-experiments/cache/openai-parameter-golf/requirements.txt zstandard
```

```python
%%bash
set -euo pipefail
cd /content/parameter-golf
python3 research-experiments/scripts/prepare_challenge_data.py --variant sp1024 --train-shards 80
```

```python
%%bash
set -euo pipefail
cd /content/parameter-golf/research-experiments
python3 scripts/run_record_experiment.py \
  --experiment-dir record_candidates/2026-03-23_rank1_mixed_qat \
  --run-id rank1_mixed_qat_record_preflight \
  --seed 42 \
  --nproc-per-node 8 \
  --stats-path runs/rank1_mixed_qat_record_preflight/stats.json \
  --preflight-only
```

```python
%%bash
set -euo pipefail
cd /content/parameter-golf/research-experiments
python3 scripts/run_record_experiment.py \
  --experiment-dir record_candidates/2026-03-23_rank1_mixed_qat \
  --run-id rank1_mixed_qat_colab_pilot_seed42 \
  --seed 42 \
  --nproc-per-node 1 \
  --required-cuda-devices 1 \
  --set-env-file record_candidates/2026-03-23_rank1_mixed_qat/colab_pilot_env.json \
  --stats-path runs/rank1_mixed_qat_colab_pilot_seed42/stats.json
```
